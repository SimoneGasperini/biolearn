```python
# import and normalize the full MNIST digits dataset

from sklearn.datasets import fetch_openml
X, y = fetch_openml(name='mnist_784', version=1, data_id=None, return_X_y=True)

X *= 1. / 255
```

```python
# define and train the BCM model

from biolearn.model.bcm import BCM
from biolearn.utils.optimizer import Adam

model = BCM(outputs=100, num_epochs=10, batch_size=2000, activation='relu',
            optimizer=Adam(lr=4e-2), interaction_strength=0., verbose=False)
model.fit(X)
```

```python
# show the receptive fields of all the neurons

from biolearn.utils.misc import view_weights

view_weights(model.weights, dims=(28, 28))
```

```python
# define and train the Hopfield model

from biolearn.model.hopfield import Hopfield

model = Hopfield(outputs=100, num_epochs=10, batch_size=100, verbose=False)
model.fit(X)
```

```python
# show the receptive fields of all the neurons

from biolearn.utils.misc import view_weights

view_weights(model.weights, dims=(28, 28))
```
